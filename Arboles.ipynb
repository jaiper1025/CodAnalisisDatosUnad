```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
data = pd.read_csv("winequality-red.csv")
data.head()
```

```python
data.shape
```

```python
data.describe()
```

```python
data.isna().sum()
```

```python
plt.hist(data.quality)
```

```python
sns.scatterplot(x = "pH", y ="alcohol", hue = "quality", data = data)
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=data[['fixed acidity',	'volatile acidity',	'citric acid',	'residual sugar',	'chlorides',	'free sulfur dioxide',	'total sulfur dioxide',	'density',	'pH',	'sulphates',	'alcohol',	'quality']])
plt.xticks(rotation=45)
plt.title('Boxplot de calidad del vino')
plt.ylabel('Valores')
plt.show()
```

```python
predictors_col=['fixed acidity',	'volatile acidity',	'citric acid',	'residual sugar',	'chlorides',	'free sulfur dioxide',	'total sulfur dioxide',	'density',	'pH',	'sulphates',	'alcohol',	'quality']
target_col=['quality']
```

```python
predictors = data[predictors_col]
target = data[target_col]
```

```python
predictors
```

```python
corr = predictors.corr()
corr
```

```python
sns.heatmap(corr)
plt.title('Matriz de Correlación')
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=0.2, random_state=13)
```

```python
tree = DecisionTreeClassifier()
```

```python
arbol = tree.fit(X_train, y_train)
```

```python
plot_tree(arbol)
```

```python
predicciones = arbol.predict(X_test)
```

```python
predicciones
```

```python
pd.crosstab(np.array([y[0] for y in y_test.values.tolist()]), predicciones, rownames=['Actual'], colnames=['Predicciones'])
```

```python
accuracy = accuracy_score(y_test,predicciones)
accuracy
```

```python
precision = precision_score(y_test, predicciones, average='weighted')
precision
```

```python
recall = recall_score(y_test, predicciones, average='weighted')
recall
```

```python
f1 = f1_score(y_test, predicciones, average='weighted')
f1
```
